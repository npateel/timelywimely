```python
import json
import re
```

```python
import gzip
from glob import glob
from tqdm import tqdm

i = 0
list_of_q = []


        
def get_answers(question):
    answers = []
    for c in question['long_answer_candidates']:
        start_token = c['start_token']
        end_token = c['end_token']
        #print(start_token, end_token)
        a = " ".join(t['token'] for t in question['document_tokens'][start_token:end_token]if not t['html_token'])
        answers.append(a)
    return answers


def get_tokens(question):
    answers = []
    for c in question['long_answer_candidates']:
        start_token = c['start_token']
        end_token = c['end_token']
        answers.append((start_token, end_token, len(question['document_tokens'])))
    return answers   


def get_short_answers(question):
    answers = []
    for annotation in question['annotations']:
        if len(annotation['short_answers']) != 0:
            short_answer = annotation['short_answers'][0]
#             print(short_answer)
            start_token = short_answer['start_token']
            end_token = short_answer['end_token']
#             print(question['document_tokens'][start_token:end_token])
            a = " ".join(t['token'] for t in question['document_tokens'][start_token:end_token])
            answers.append(a)
#         else:
#             print('No short answers found in annotation.')
    return list(set(answers))
```

```python
# CHANGE THIS FILE NAME
finalList = []
count = 0
with open('./temp.test.jsonl', 'r') as json_file:
        json_list = list(json_file)
nqot = {}
for json_str in json_list:
    result = json.loads(json_str)
    q = re.sub("\?", "", result["question"])
    nqot[q] = (result["is_dependent"] == "yes")
print(nqot)
for file in glob('../../Research/nq/v1.0/dev/nq-dev-*.jsonl.gz'):
    list_of_q = []
    with gzip.open(file) as f:
        for line in tqdm(f):
            list_of_q.append(line)
            i += 1

    ind = []
    exs = []
    for i in tqdm(range(len(list_of_q))):
        l = list_of_q[i]
        x = json.loads(l)
        short_ans = get_short_answers(x)
        if len(short_ans) != 0:
            ind.append(i)
            exs.append(x)

    nq_qs = []
    for x in tqdm(exs):
        #x = json.loads(list_of_q[i])
        #ls = get_answers(x)
        q = x["question_text"]
        sa = get_short_answers(x)
        title = x["document_title"]
        nq_qs.append({"question": q, "answers": sa, "topic": title})
    nq = {"data": nq_qs}
    nqdict = {}
    i = 0
    for q in nq["data"]:
        nqdict[q["question"]] = i;
        i += 1;
    #print(q["question"])
    #print(nqdict)

    
        
    print(len(nqot), len(nqdict))
    for q in nqot:
        if q in nqdict:
            count += 1
            nq["data"][nqdict[q]]["in_dependent"] = nqot[q]
            finalList.append(nq["data"][nqdict[q]])
    print(count)
        
finalDict = {"data": finalList}
with open('nq_open_topics.json', 'w') as fp:
    json.dump(finalDict, fp)
```
